```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,10)
y = np.sin(x)
z = np.cos(x)

plt.plot(x,y,'b',x,z,'r')
plt.xlabel('Radians');
plt.ylabel('value');
plt.title('ploting demonstration')
plt.legend(['sin','cos'])
plt.grid()
```

```python
harga_awal = 650000 
jumlah_awal = 25     
total_invest_awal = harga_awal * jumlah_awal

harga_saat_ini = 685000
nilai_saat_ini = harga_saat_ini * jumlah_awal

profit_awal = nilai_saat_ini - total_invest_awal
profit_persen_awal = (profit_awal / total_invest_awal) * 100

print("Keuntungan dari pembelian awal:")
print("Dalam Rupiah: Rp", profit_awal)
print("Dalam Persen:", round(profit_persen_awal, 2), "%")

jumlah_tambah = 15
harga_tambah = harga_saat_ini
total_invest_tambah = harga_tambah * jumlah_tambah

total_gram = jumlah_awal + jumlah_tambah
total_invest_total = total_invest_awal + total_invest_tambah

harga_akhir = 715000
nilai_akhir = total_gram * harga_akhir

profit_total = nilai_akhir - total_invest_total
profit_persen_total = (profit_total / total_invest_total) * 100

print("\nKeuntungan setelah pembelian tambahan dan kenaikan harga:")
print("Dalam Rupiah: Rp", profit_total)
print("Dalam Persen:", round(profit_persen_total, 2), "%")
```

```python
import math
uang_awal = 200000000
target_uang = 400000000
bunga_tahun = 10/100

waktu_butuh = math.log(target_uang / uang_awal) / (math.log(1 + bunga_tahun))

print ("waktu yang di butuh kan adalah : ",(round(waktu_butuh)), "tahun")
```
